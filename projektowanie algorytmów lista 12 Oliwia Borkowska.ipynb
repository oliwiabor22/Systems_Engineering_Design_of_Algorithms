```python
# projektowanie algorytmów Oliwia Borkowska lista 12 - ścieżki i przepływy

# zadanie 1 

# 1 - Zaimplementuj strukturę danych do przechowywania drzewa binarnego – każdy wierzchołek ma najwyżej dwóch potomków. Przechowuj tylko wierzchołki i strukturę połączeń między nimi.

import csv

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

# 2 - Zaimplementuj funkcję dodającą wierzchołek zadany przez użytkownika, w miejscu zadanym przezużytkownika. Konkretnie, wierzchołek może być dodany w miejscu jeszcze niezajętego potomka dowolnego wierzchołka drzewa
    
    def find_node(self, current, value):
        if current is None:
            return None
        if current.value == value:
            return current
        return self.find_node(current.left, value) or self.find_node(current.right, value)

    def add_node_user_input(self):
        if self.root is None:
            val = input("Drzewo jest puste. Podaj wartość korzenia: ")
            self.root = Node(val)
            print(f"Ustawiono {val} jako korzeń.")
            return

        parent_val = input("Podaj wartość węzła, do którego chcesz dodać potomka: ")
        parent = self.find_node(self.root, parent_val)
        if not parent:
            print("Nie znaleziono takiego węzła.")
            return

        new_val = input("Podaj wartość nowego węzła: ")
        side = input("Gdzie dodać potomka? (left/right): ").strip().lower()

        if side == "left":
            if parent.left is None:
                parent.left = Node(new_val)
                print(f"Dodano {new_val} jako lewego potomka {parent_val}.")
            else:
                print("Lewy potomek już istnieje.")
        elif side == "right":
            if parent.right is None:
                parent.right = Node(new_val)
                print(f"Dodano {new_val} jako prawego potomka {parent_val}.")
            else:
                print("Prawy potomek już istnieje.")
        else:
            print("Nieprawidłowy kierunek. Użyj 'left' lub 'right'.")

# 3 - Zaimplementuj funkcję usuwającą wierzchołek wskazany przez użytkownika. W przypadku usuwania wierzchołka wewnętrznego, usuń całe poddrzewo
    def delete_node_user_input(self):
        if self.root is None:
            print("Drzewo jest puste.")
            return

        target = input("Podaj wartość węzła do usunięcia (z jego poddrzewem): ")

        if self.root.value == target:
            self.root = None
            print("Usunięto cały korzeń drzewa.")
            return

        success = self._delete_node(self.root, target)
        if success:
            print(f"Węzeł {target} został usunięty.")
        else:
            print("Nie znaleziono węzła.")

    def _delete_node(self, current, target):
        if current is None:
            return False
        if current.left and current.left.value == target:
            current.left = None
            return True
        if current.right and current.right.value == target:
            current.right = None
            return True
        return self._delete_node(current.left, target) or self._delete_node(current.right, target)

# 4 - Zaimplementuj funkcję graficznie wyświetlającą przechowywany graf
    def display(self, node=None, level=0, prefix="Root: "):
        if node is None:
            node = self.root
        if node is not None:
            print(" " * (4 * level) + prefix + str(node.value))
            if node.left:
                self.display(node.left, level + 1, "L--- ")
            if node.right:
                self.display(node.right, level + 1, "R--- ")

# 5 - Zaimplementuj funkcję zapisującą drzewo do pliku  - PLIK CSV
    def save_to_file(self, filename):
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['parent', 'left', 'right'])
            self._write_node(writer, self.root)

    def _write_node(self, writer, node):
        if node is None:
            return
        left_val = node.left.value if node.left else ''
        right_val = node.right.value if node.right else ''
        writer.writerow([node.value, left_val, right_val])
        self._write_node(writer, node.left)
        self._write_node(writer, node.right)

# 5 - wczytującą drzewo z pliku. Zapisuj wierzchołki i połączenia. Zaproponuj format pliku.
    def load_from_file(self, filename):
        nodes = {}
        connections = []

        with open(filename, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                parent = row['parent']
                left = row['left']
                right = row['right']
                if parent not in nodes:
                    nodes[parent] = Node(parent)
                if left and left not in nodes:
                    nodes[left] = Node(left)
                if right and right not in nodes:
                    nodes[right] = Node(right)
                connections.append((parent, left, right))

        for parent, left, right in connections:
            node = nodes[parent]
            node.left = nodes.get(left)
            node.right = nodes.get(right)

        self.root = nodes[connections[0][0]] if connections else None
        print(f"Wczytano drzewo z pliku '{filename}'.")


if __name__ == "__main__":
    tree = BinaryTree()

    while True:
        print("\n--- MENU ---")
        print("1. Dodaj węzeł")
        print("2. Usuń węzeł")
        print("3. Wyświetl drzewo")
        print("4. Zapisz drzewo do pliku")
        print("5. Wczytaj drzewo z pliku")
        print("6. Zakończ")

        opcja = input("Wybierz opcję (1-6): ").strip()

        if opcja == "1":
            tree.add_node_user_input()
        elif opcja == "2":
            tree.delete_node_user_input()
        elif opcja == "3":
            tree.display()
        elif opcja == "4":
            tree.save_to_file("drzewo1.csv")
        elif opcja == "5":
            tree.load_from_file("drzewo1.csv")
        elif opcja == "6":
            print("Zakończono program.")
            break
        else:
            print("Nieprawidłowa opcja.")
```

```python

# tak jak w liście 8 - lista robotów:

import random
import csv

TYP = ["AGV", "AFV", "ASV", "AUV"]

def generuj_losowego_robota():
    typ = random.choice(TYP)
    cena = round(random.uniform(0, 10000), 2)
    zasieg = random.randint(0, 100)
    kamera = random.choice([0, 1])
    return (typ, cena, zasieg, kamera)

def generuj_liste_robotow(N):
    lista_robotow = []
    for _ in range(N):
        lista_robotow.append(generuj_losowego_robota())
    return lista_robotow

def wyswietl_liste_robotow(lista_robotow):
    print(f"{'TYP':<5} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 40)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<5} {cena:>10.2f} {zasieg:>10} {kamera:>10}")

def zapisz_do_pliku(lista_robotow, nazwa_pliku):
    with open(nazwa_pliku, "w", newline='') as plik:
        writer = csv.writer(plik)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        writer.writerows(lista_robotow)

def odczytaj_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

if __name__ == "__main__":
    N = int(input("Podaj liczbę robotów do wygenerowania i zapisania na liście: "))
    lista = generuj_liste_robotow(N)

    print("Lista robotów:")
    wyswietl_liste_robotow(lista)

    nazwa_pliku = "lista_robotow.csv"
    zapisz_do_pliku(lista, nazwa_pliku)
    print(f"\n\nLista robotów została zapisana do pliku '{nazwa_pliku}'.")

    print(f"\n\nOdczytana lista robotów z pliku '{nazwa_pliku}':")
    lista_z_pliku = odczytaj_z_pliku(nazwa_pliku)
    wyswietl_liste_robotow(lista_z_pliku)
```

```python
import csv
import random

class Robot:
    def __init__(self, typ, cena, zasieg, kamera):
        self.typ = typ
        self.cena = cena
        self.zasieg = zasieg
        self.kamera = kamera

    def __str__(self):
        return f"{self.typ} | Cena: {self.cena} | Zasięg: {self.zasieg} | Kamera: {self.kamera}"

class BSTNode:
    def __init__(self, robot, key_attr):
        self.robot = robot
        self.key_attr = key_attr
        self.key = getattr(robot, key_attr)
        self.left = None
        self.right = None

class BST:
    def __init__(self, key_attr):
        self.root = None
        self.key_attr = key_attr

    def insert(self, robot):
        self.root = self._insert(self.root, robot)

    def _insert(self, node, robot):
        if node is None:
            return BSTNode(robot, self.key_attr)
        if getattr(robot, self.key_attr) < node.key:
            node.left = self._insert(node.left, robot)
        else:
            node.right = self._insert(node.right, robot)
        return node

    def delete(self, key_value):
        self.root = self._delete(self.root, key_value)

    def _delete(self, node, key_value):
        if node is None:
            return None
        if key_value < node.key:
            node.left = self._delete(node.left, key_value)
        elif key_value > node.key:
            node.right = self._delete(node.right, key_value)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            min_larger_node = self._find_min(node.right)
            node.robot = min_larger_node.robot
            node.key = min_larger_node.key
            node.right = self._delete(node.right, min_larger_node.key)
        return node

    def _find_min(self, node):
        while node.left:
            node = node.left
        return node

    def search(self, key_value):
        return self._search(self.root, key_value)

    def _search(self, node, key_value):
        if node is None:
            return None
        if key_value == node.key:
            return node.robot
        elif key_value < node.key:
            return self._search(node.left, key_value)
        else:
            return self._search(node.right, key_value)

    def display(self):
        lines, *_ = self._display_aux(self.root)
        for line in lines:
            print(line)

    def _display_aux(self, node):
        if node is None:
            return ["<puste>"], 0, 0, 0

        line = str(node.robot)
        width = len(line)
        height = 1
        middle = width // 2

        if node.left is None and node.right is None:
            return [line], width, height, middle

        if node.left is None:
            lines, n, p, x = self._display_aux(node.right)
            first_line = line + ' ' * (n + 1)
            second_line = ' ' * middle + '\\' + ' ' * (n + 1)
            shifted_lines = [l + ' ' * (width + 1) for l in lines]
            return [first_line, second_line] + shifted_lines, width + n + 1, p + 2, middle

        if node.right is None:
            lines, n, p, x = self._display_aux(node.left)
            first_line = ' ' * (n + 1) + line
            second_line = ' ' * (n + 1) + '/' + ' ' * middle
            shifted_lines = [' ' * (width + 1) + l for l in lines]
            return [first_line, second_line] + shifted_lines, n + width + 1, p + 2, n + 1 + middle

        left, n, p, x = self._display_aux(node.left)
        right, m, q, y = self._display_aux(node.right)
        s = line
        u = len(s)
        first_line = (" " * (x + 1)) + ("_" * (n - x - 1)) + s + ("_" * y) + (" " * (m - y))
        second_line = (" " * x + "/") + (" " * (n - x - 1 + u + y)) + "\\" + (" " * (m - y - 1))
        if p < q:
            left += [" " * n] * (q - p)
        elif q < p:
            right += [" " * m] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + " " * u + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    def save_to_file(self, filename):
        with open(filename, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
            self._save(self.root, writer)

    def _save(self, node, writer):
        if node:
            r = node.robot
            writer.writerow([r.typ, r.cena, r.zasieg, r.kamera])
            self._save(node.left, writer)
            self._save(node.right, writer)

    def load_from_file(self, filename):
        with open(filename, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                robot = Robot(row['TYP'], float(row['CENA']), int(row['ZASIEG']), int(row['KAMERA']))
                self.insert(robot)

def generuj_roboty(n):
    TYP = ["AGV", "AFV", "ASV", "AUV"]
    return [Robot(random.choice(TYP), round(random.uniform(0, 10000), 2), random.randint(0, 100), random.choice([0, 1])) for _ in range(n)]

def zapisz_roboty_do_pliku(roboty, plik):
    with open(plik, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        for r in roboty:
            writer.writerow([r.typ, r.cena, r.zasieg, r.kamera])

def wczytaj_roboty_z_pliku(plik):
    roboty = []
    with open(plik, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            roboty.append(Robot(row['TYP'], float(row['CENA']), int(row['ZASIEG']), int(row['KAMERA'])))
    return roboty

if __name__ == '__main__':
    plik = 'roboty.csv'
    roboty = generuj_roboty(10)
    zapisz_roboty_do_pliku(roboty, plik)

    print("Wczytano roboty z pliku:")
    roboty = wczytaj_roboty_z_pliku(plik)
    for r in roboty:
        print(r)

    klucz = input("\nPodaj parametr (typ, cena, zasieg, kamera), wg ktorego chcesz utworzyc BST: ").strip().lower()
    bst = BST(klucz)
    for r in roboty:
        bst.insert(r)

    print("\nDrzewo BST:")
    bst.display()

    print("\nDodaj robota:")
    typ = input("TYP: ")
    cena = float(input("CENA: "))
    zasieg = int(input("ZASIEG: "))
    kamera = int(input("KAMERA (0/1): "))
    bst.insert(Robot(typ, cena, zasieg, kamera))
    print("\nPo dodaniu:")
    bst.display()

    key_value = input(f"\nPodaj wartosc klucza ({klucz}) do usuniecia: ")
    if klucz in ["cena"]:
        key_value = float(key_value)
    elif klucz in ["zasieg", "kamera"]:
        key_value = int(key_value)
    bst.delete(key_value)
    print("\nPo usunieciu:")
    bst.display()

    print("\nZapis drzewa do pliku 'bst_out.csv'")
    bst.save_to_file("bst_out.csv")

    print("\nOdczyt z pliku 'bst_out.csv':")
    nowy_bst = BST(klucz)
    nowy_bst.load_from_file("bst_out.csv")
    nowy_bst.display()
```

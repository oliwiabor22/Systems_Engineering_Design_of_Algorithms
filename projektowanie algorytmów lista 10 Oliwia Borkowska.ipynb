```python
# projektowanie algorytmów lista 10 - algorytmy sortowania Oliwia Borkowska

# tak jak w liście 8 i 9 - lista robotów:

import random
import csv

TYP = ["AGV", "AFV", "ASV", "AUV"]

def generuj_losowego_robota():
    typ = random.choice(TYP)
    cena = round(random.uniform(0, 10000), 2)
    zasieg = random.randint(0, 100)
    kamera = random.choice([0, 1])
    return (typ, cena, zasieg, kamera)

def generuj_liste_robotow(N):
    lista_robotow = []
    for _ in range(N):
        lista_robotow.append(generuj_losowego_robota())
    return lista_robotow

def wyswietl_liste_robotow(lista_robotow):
    print(f"{'TYP':<5} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 40)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<5} {cena:>10.2f} {zasieg:>10} {kamera:>10}")

def zapisz_do_pliku(lista_robotow, nazwa_pliku):
    with open(nazwa_pliku, "w", newline='') as plik:
        writer = csv.writer(plik)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        writer.writerows(lista_robotow)

def odczytaj_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

if __name__ == "__main__":
    N = int(input("Podaj liczbę robotów do wygenerowania i zapisania na liście: "))
    lista = generuj_liste_robotow(N)

    print("Lista robotów:")
    wyswietl_liste_robotow(lista)

    nazwa_pliku = "lista_robotow.csv"
    zapisz_do_pliku(lista, nazwa_pliku)
    print(f"\n\nLista robotów została zapisana do pliku '{nazwa_pliku}'.")

    print(f"\n\nOdczytana lista robotów z pliku '{nazwa_pliku}':")
    lista_z_pliku = odczytaj_z_pliku(nazwa_pliku)
    wyswietl_liste_robotow(lista_z_pliku)
```

```python
# zadanie 1 - sortowanie przez kopcowanie

import csv

def odczytanie_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)  # pomijamy nagłówek
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

def zapisanie_do_pliku(lista_robotow, nazwa_pliku):
    with open(nazwa_pliku, "w", newline='') as plik:
        writer = csv.writer(plik)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        writer.writerows(lista_robotow)

def wyswietlenie_listy_robotow(lista_robotow):
    print(f"{'TYP':<5} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 40)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<5} {cena:>10.2f} {zasieg:>10} {kamera:>10}")
    print("-" * 40)

def heapify(lista, n, i, krokowo=False):
    largest = i
    lewy = 2 * i + 1
    prawy = 2 * i + 2

    if lewy < n and lista[lewy][1] > lista[largest][1]:
        largest = lewy

    if prawy < n and lista[prawy][1] > lista[largest][1]:
        largest = prawy

    if largest != i:
        lista[i], lista[largest] = lista[largest], lista[i]
        if krokowo:
            print(f"Po zamianie elementów indeks {i} i {largest}:")
            wyswietlenie_listy_robotow(lista)
        heapify(lista, n, largest, krokowo)

def heap_sort(lista, krokowo=False):
    n = len(lista)

    # stworzenie kopca maksymalnego
    for i in range(n // 2 - 1, -1, -1):
        heapify(lista, n, i, krokowo)
        if krokowo:
            print(f"Po heapify({i}):")
            wyswietl_liste_robotow(lista)

    for i in range(n - 1, 0, -1):
        lista[i], lista[0] = lista[0], lista[i]
        if krokowo:
            print(f"\nPo zamianie korzenia z elementem indeks {i}:")
            wyswietlenie_listy_robotow(lista)
        heapify(lista, i, 0, krokowo)

if __name__ == "__main__":
    nazwa_pliku_wejsciowego = input("Podaj nazwę pliku CSV z listą robotów do wczytania: ")
    roboty = odczytanie_z_pliku(nazwa_pliku_wejsciowego)

    print("\nOryginalna lista robotów:")
    wyswietlenie_listy_robotow(roboty)

    print("\nSortowanie przez kopcowanie (krok-po-kroku)")
    heap_sort(roboty, krokowo=True)

    print("\nPosortowana lista robotów względem ceny (rosnąco)")
    wyswietlenie_listy_robotow(roboty)

    plik_wyjsciowy = "posortowana_lista_robotow.csv"
    zapisanie_do_pliku(roboty, plik_wyjsciowy)
    print(f"\nPosortowana lista robotów zapisana w: {plik_wyjsciowy}")
```

```python
# zadanie 2 - sortowanie szybkie

import csv

def odczytanie_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)  # pomijamy nagłówek
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

def zapisanie_do_pliku(lista_robotow, nazwa_pliku):
    with open(nazwa_pliku, "w", newline='') as plik:
        writer = csv.writer(plik)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        writer.writerows(lista_robotow)

def wyswietlenie_listy_robotow(lista_robotow):
    print(f"{'TYP':<5} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 40)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<5} {cena:>10.2f} {zasieg:>10} {kamera:>10}")
    print("-" * 40)

def wyswietlenie_ceny(roboty):
    ceny = [f"{robot[1]:.2f}" for robot in roboty]
    print("Ceny:", ceny)

def quicksort(roboty, low, high, krokowo=False):
    if low < high:
        pi = partition(roboty, low, high, krokowo)
        if krokowo:
            print(f"Po podziale (pivot={roboty[pi][1]:.2f}):")
            wyswietlenie_ceny(roboty)
        quicksort(roboty, low, pi - 1, krokowo)
        quicksort(roboty, pi + 1, high, krokowo)

def partition(roboty, low, high, krokowo):
    pivot = roboty[high][1]
    i = low - 1

    for j in range(low, high):
        if roboty[j][1] <= pivot:
            i += 1
            roboty[i], roboty[j] = roboty[j], roboty[i]

    roboty[i + 1], roboty[high] = roboty[high], roboty[i + 1]
    return i + 1

if __name__ == "__main__":
    nazwa_pliku_wejsciowego = input("Podaj nazwę pliku CSV z listą robotów do wczytania: ")
    roboty = odczytaj_z_pliku(nazwa_pliku_wejsciowego)

    print("\nOryginalna lista robotów:")
    wyswietl_liste_robotow(roboty)

    print("\nSortowanie szybkie (krok-po-kroku)")
    quicksort(roboty, 0, len(roboty) - 1, krokowo=True)

    print("\nosortowana lista robotów względem ceny (rosnąco)")
    wyswietl_liste_robotow(roboty)

    plik_wyjsciowy2 = "posortowana_quicksort_lista_robotow.csv"
    zapisz_do_pliku(roboty, plik_wyjsciowy2)
    print(f"\nPosortowana lista zapisana w: {plik_wyjsciowy2}")
```

```python
# zadanie 3 – sortowanie przez zliczanie

import csv

def odczytanie_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)  # pomijamy nagłówek
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

def zapisanie_do_pliku(lista_robotow, nazwa_pliku):
    with open(nazwa_pliku, "w", newline='') as plik:
        writer = csv.writer(plik)
        writer.writerow(["TYP", "CENA", "ZASIEG", "KAMERA"])
        writer.writerows(lista_robotow)

def wyswietlenie_listy_robotow(lista_robotow):
    print(f"{'TYP':<5} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 40)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<5} {cena:>10.2f} {zasieg:>10} {kamera:>10}")
    print("-" * 40)

def counting_sort(roboty, maks_zasieg=100):
    count = [[] for _ in range(maks_zasieg + 1)] # zasięg od 0–maks_zasieg

    for robot in roboty:
        zasieg = robot[2]
        count[zasieg].append(robot)

    posortowane = []
    for podlista in count:
        posortowane.extend(podlista)

    return posortowane

if __name__ == "__main__":
    nazwa_pliku_wejsciowego = input("Podaj nazwę pliku CSV z listą robotów do wczytania: ")
    roboty = odczytanie_z_pliku(nazwa_pliku_wejsciowego)

    print("\nOryginalna lista robotów:")
    wyswietlenie_listy_robotow(roboty)

    print("\nSortowanie przez zliczanie względem zasięgu")
    roboty_posortowane = counting_sort(roboty)

    print("\nPosortowana lista robotów względem zasięgu (rosnąco)")
    wyswietlenie_listy_robotow(roboty_posortowane)

    plik_wyjsciowy3 = "posortowana_zasiegiem_lista_robotow.csv"
    zapisz_do_pliku(roboty_posortowane, plik_wyjsciowy3)
    print(f"\nPosortowana lista została zapisana do pliku: {plik_wyjsciowy3}")
```

```python
# zadanie 4 - sortowanie pozycyjne
def wczytanie_macierzy():
    m = int(input("Podaj liczbę wierszy (M): "))
    n = int(input("Podaj liczbę kolumn (N): "))

    print(f"Podaj kolejne {m} wierszy po {n} liczb całkowitych oddzielonych spacją:")
    macierz = []
    for i in range(m):
        while True:
            try:
                wiersz = list(map(int, input(f"Wiersz {i+1}: ").split()))
                if len(wiersz) != n:
                    raise ValueError("Nieprawidłowa liczba elementów.")
                macierz.append(wiersz)
                break
            except ValueError as e:
                print(f"Błąd: {e}. Spróbuj ponownie.")
    return macierz

def wyswietlenie_macierzy(macierz, naglowek="MACIERZ:"):
    print("\n" + naglowek)
    for wiersz in macierz:
        print(" ".join(f"{el:4d}" for el in wiersz))
    print()

def sortowanie_pozycyjne(macierz):
    return sorted(macierz)

if __name__ == "__main__":
    macierz = wczytanie_macierzy()
    wyswietlenie_macierzy(macierz, "ORGINALNA MACIERZ:")

    posortowana = sortowanie_pozycyjne(macierz)
    wyswietlenie_macierzy(posortowana, "MACIERZ POSORTOWANA POZYCYJNIE:")
```

```python
# zadanie 5 - zestawienie algorytmów

import csv
import matplotlib.pyplot as plt

def odczytanie_z_pliku(nazwa_pliku):
    lista_robotow = []
    with open(nazwa_pliku, "r", newline='') as plik:
        reader = csv.reader(plik)
        next(reader)
        for wiersz in reader:
            typ = wiersz[0]
            cena = float(wiersz[1])
            zasieg = int(wiersz[2])
            kamera = int(wiersz[3])
            lista_robotow.append((typ, cena, zasieg, kamera))
    return lista_robotow

def wyswietlenie_listy_robotow(lista_robotow):
    print(f"{'TYP':<10} {'CENA':>10} {'ZASIĘG':>10} {'KAMERA':>10}")
    print("-" * 45)
    for robot in lista_robotow:
        typ, cena, zasieg, kamera = robot
        print(f"{typ:<10} {cena:>10.2f} {zasieg:>10} {kamera:>10}")
    print("-" * 45)

def heapify(lista, n, i, steps):
    largest = i
    lewy = 2 * i + 1
    prawy = 2 * i + 2

    if lewy < n and lista[lewy][1] > lista[largest][1]:
        largest = lewy

    if prawy < n and lista[prawy][1] > lista[largest][1]:
        largest = prawy

    if largest != i:
        lista[i], lista[largest] = lista[largest], lista[i]
        steps.append([x[1] for x in lista])
        heapify(lista, n, largest, steps)

def heap_sort(lista):
    n = len(lista)
    steps = []
    for i in range(n // 2 - 1, -1, -1):
        heapify(lista, n, i, steps)
    for i in range(n - 1, 0, -1):
        lista[i], lista[0] = lista[0], lista[i]
        steps.append([x[1] for x in lista])
        heapify(lista, i, 0, steps)
    return steps


def partition(roboty, low, high, steps):
    pivot = roboty[high][1]
    i = low - 1
    for j in range(low, high):
        if roboty[j][1] <= pivot:
            i += 1
            roboty[i], roboty[j] = roboty[j], roboty[i]
            steps.append([x[1] for x in roboty])
    roboty[i + 1], roboty[high] = roboty[high], roboty[i + 1]
    steps.append([x[1] for x in roboty])
    return i + 1

def quicksort(roboty, low, high, steps):
    if low < high:
        pi = partition(roboty, low, high, steps)
        quicksort(roboty, low, pi - 1, steps)
        quicksort(roboty, pi + 1, high, steps)

def quick_sort_wrapper(lista):
    steps = []
    quicksort(lista, 0, len(lista) - 1, steps)
    return steps


def rysuj_wykresy(kroki, tytul, max_wykresow=6, co_krok=5):
    n = len(kroki)
    kroki_do_rysowania = [0] + list(range(0, n, co_krok))
    if kroki_do_rysowania[-1] != n-1:
        kroki_do_rysowania.append(n-1)
    kroki_do_rysowania = kroki_do_rysowania[:max_wykresow]

    fig, axes = plt.subplots(1, len(kroki_do_rysowania), figsize=(4 * len(kroki_do_rysowania), 4))
    if len(kroki_do_rysowania) == 1:
        axes = [axes]
    for ax, idx in zip(axes, kroki_do_rysowania):
        ax.bar(range(len(kroki[idx])), kroki[idx], color='skyblue')
        ax.set_title(f"{tytul}\nKrok {idx+1}/{n}")
        ax.set_xlabel('Indeksy')
        ax.set_ylabel('Cena')
        ax.set_ylim(0, max(max(kroki[-1]), max(kroki[idx])) * 1.1)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    nazwa_pliku = input("Podaj nazwę pliku CSV z listą robotów do wczytania: ")
    roboty_heap = odczytanie_z_pliku(nazwa_pliku)
    roboty_quick = roboty_heap.copy()

    print("\nOryginalna lista robotów:")
    wyswietlenie_listy_robotow(roboty_heap)

    print("\nSortowanie przez kopcowanie - generowanie kroków...")
    heap_steps = heap_sort(roboty_heap)

    print("\nSortowanie szybkie - generowanie kroków...")
    quick_steps = quick_sort_wrapper(roboty_quick)

    print("\nRysowanie wykresów dla sortowania przez kopcowanie...")
    rysuj_wykresy(heap_steps, "Sortowanie przez kopcowanie")

    print("\nRysowanie wykresów dla sortowania szybkiego...")
    rysuj_wykresy(quick_steps, "Sortowanie szybkie")
```

```python
import csv
import copy
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib notebook


def wczytaj_roboty_z_pliku(sciezka):
    roboty = []
    with open(sciezka, newline='', encoding='utf-8') as plik:
        reader = csv.DictReader(plik)
        for wiersz in reader:
            typ = wiersz['TYP']
            cena = float(wiersz['CENA'])
            zasieg = int(wiersz['ZASIEG'])
            kamera = int(wiersz['KAMERA'])
            robot = (typ, cena, zasieg, kamera)
            roboty.append(robot)
    return roboty

def zapisz_roboty_do_pliku(roboty, sciezka):
    with open(sciezka, 'w', newline='', encoding='utf-8') as plik:
        fieldnames = ['TYP', 'CENA', 'ZASIEG', 'KAMERA']
        writer = csv.DictWriter(plik, fieldnames=fieldnames)
        writer.writeheader()
        for r in roboty:
            writer.writerow({'TYP': r[0], 'CENA': r[1], 'ZASIEG': r[2], 'KAMERA': r[3]})

def przywroc_kopiec(arr, n, i, steps):
    najwiekszy = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n:
        if arr[l][1] > arr[najwiekszy][1]:
            najwiekszy = l

    if r < n:
        if arr[r][1] > arr[najwiekszy][1]:
            najwiekszy = r

    if najwiekszy != i:
        arr[i], arr[najwiekszy] = arr[najwiekszy], arr[i]
        steps.append([r[1] for r in arr])
        przywroc_kopiec(arr, n, najwiekszy, steps)
    else:
        steps.append([r[1] for r in arr])


def heapsort(arr):
    steps = []
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        przywroc_kopiec(arr, n, i, steps)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps.append([r[1] for r in arr])
        przywroc_kopiec(arr, i, 0, steps)

    return steps

def quicksort(arr, low, high, steps):
    if low < high:
        pi = podziel(arr, low, high, steps)
        quicksort(arr, low, pi - 1, steps)
        quicksort(arr, pi + 1, high, steps)

def podziel(arr, low, high, steps):
    pivot = arr[high][1]
    i = low - 1
    for j in range(low, high):
        if arr[j][1] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            steps.append([r[1] for r in arr])
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    steps.append([r[1] for r in arr])
    return i + 1


def main():
    nazwa_pliku = input("Podaj nazwę pliku CSV z listą robotów (np. lista_robotow.csv): ").strip()
    if not nazwa_pliku:
        nazwa_pliku = "lista_robotow.csv"

    roboty = wczytaj_roboty_z_pliku(nazwa_pliku)

    roboty_heap = copy.deepcopy(roboty)
    roboty_quick = copy.deepcopy(roboty)

    kroki_heap = heapsort(roboty_heap)
    kroki_quick = []
    quicksort(roboty_quick, 0, len(roboty_quick) - 1, kroki_quick)

    zapisz_roboty_do_pliku(roboty_heap, "posortowana_lista_robotow.csv")
    zapisz_roboty_do_pliku(roboty_quick, "posortowana_quicksort_lista_robotow.csv")

    print("Liczba kroków heapsort:", len(kroki_heap))
    print("Liczba kroków quicksort:", len(kroki_quick))

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.set_title("Sortowanie przez kopcowanie")
    ax2.set_title("Sortowanie szybkie")

    max_cena = max(r[1] for r in roboty)
    ax1.set_ylim(0, max_cena * 1.1)
    ax2.set_ylim(0, max_cena * 1.1)

    bar_rects_heap = ax1.bar(range(len(roboty)), [r[1] for r in roboty], color='violet', edgecolor='black')
    bar_rects_quick = ax2.bar(range(len(roboty)), [r[1] for r in roboty], color='violet', edgecolor='black')

    def update(frame):
        print("Wyświetlam klatkę:", frame)
        if frame < len(kroki_heap):
            for rect, h in zip(bar_rects_heap, kroki_heap[frame]):
                rect.set_height(h)
        if frame < len(kroki_quick):
            for rect, h in zip(bar_rects_quick, kroki_quick[frame]):
                rect.set_height(h)
        return bar_rects_heap + bar_rects_quick

    max_frames = max(len(kroki_heap), len(kroki_quick))
    anim = animation.FuncAnimation(fig, update, frames=max_frames, interval=500, blit=False, repeat=False)
    plt.show()
    return anim

if __name__ == "_main_":
    anim = main()
```
